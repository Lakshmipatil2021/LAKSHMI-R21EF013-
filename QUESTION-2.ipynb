```python
sns.scatterplot(data=df, x='MedInc', y='SalePrice', hue='AveRooms')
plt.show()
```

```python
cat_var = ['MedInc','HouseAge','AveRooms','AveBedrms']
num_var = ['Latitude','SalePrice']
```

```python
for column in cat_var:   
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    ax = sns.countplot(x=column, data=df, palette='Accent')
    ax.bar_label(ax.containers[0])
    plt.show()
```

```python
for column in cat_var:
    plt.subplot(1,2,2)
    ax = sns.countplot(x=column, data=df, hue='MedInc', palette='Accent')
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    plt.show()
```
