```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/drkir/Downloads/Student_Enrollment_for_Regular_courses-2015.csv")
```

```python
df
```

```python
plt.figure(figsize=(20, 8))
sns.countplot(data=df, x='levell')
plt.title('Number of students')
plt.xlabel('levell')
plt.ylabel('programme_id')
plt.xticks(rotation=90)
plt.show()
```

```python
df['programme_id'].value_counts().plot(kind='bar')
plt.title('Number of students')
plt.xlabel('University_name')
plt.ylabel('programme_id')
sns.despine()
```

```python
df['levell'].value_counts().plot(kind='bar', figsize = (14,7))
```

```python
plt.figure(figsize=(18, 9))
sns.countplot(data=df, x='levell', hue='programme')
plt.xlabel('levell')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='levell', labels=['No', 'Yes'])
plt.show()
```
